```python
import numpy as np
import pandas as pd
import datetime as dt
```

```python
data = pd.read_excel("hist_data.xlsm", sheet_name = 'SofrCurve', index_col = 'Tenor')
data.columns = [col.strftime('%Y-%m-%d') if isinstance(col, dt.datetime) else col for col in data.columns]
data
```

```python
data1 = data[["T","2023-10-30"]].copy()
data1['df'] = 1/((1+data1["2023-10-30"])**data1["T"])
data2 = data1[(data1["T"] >= 1) & (data1["T"] <= 10)].copy()
data2
```

```python
flt = 1 - data2["df"].iloc[-1]
flt
```

```python
fixed = sum(data2['df'])*0.042
fixed
```

```python
v_sofr = (flt-fixed)*1e6
v_sofr
```
